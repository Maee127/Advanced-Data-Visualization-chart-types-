### 1- 3D Scatter:

```python

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = [iris.target_names[i] for i in iris.target]

fig = px.scatter_3d(
    df, x='sepal length (cm)', y='sepal width (cm)', z='petal length (cm)',
    color='species', title='🌸 3 Species Clearly Separated in 3D Space',
    color_discrete_map={'setosa': '#FF6B6B', 'versicolor': '#4ECDC4', 'virginica': "#94D145"}
)
fig.update_layout(height=600, width=900, margin=dict(l=0, r=0, b=0, t=40))
fig.show()
```

### 2- Histogram:

```python
# PX.histogram
fig = px.histogram(
    df,
    x='sepal length (cm)',
    color='species',      # Color Bars with Species
    nbins=20,             # Number of bins in Histogram
    marginal='box',       # Adds a boxplot on the side
    title='Histogram of Sepal Width Distribution',
    labels={
        'sepal length (cm)': 'Sepal Length (cm)'
        },
    color_discrete_map={
        'setosa': '#FF6B6B',
        'versicolor':'#4ECDC4',
        'virginica':"#89D145"
    },
    opacity=0.75,
    )
fig.update_layout(
    xaxis_title='Sepal Length (cm)',
    yaxis_title='count',
    bargap=0.05 # Small gap between bars
)
fig.show()

```

### 3- Pie / Donut PLot:

with PX

```python
# Pie Chart of Sepal Width by Species

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = [iris.target_names[i] for i in iris.target]

# Create donut chart
fig = px.pie(
    df,
    names='species',
    title='Iris Species Distribution',
    color='species',
    color_discrete_map={
        'setosa': '#FF6B6B',
        'versicolor': '#4ECDC4',
        'virginica': "#94D145"
    }
)

fig.update_traces(
    textinfo='label+percent',
    textfont_size=14,
    marker=dict(line=dict(color='white', width=2))
)

fig.update_layout(
    legend_title='Species',
    width=600,
    height=600
)

fig.show()  
```

```python
# Donut Chart of Iris Species Distribution 

fig = px.pie(
    df,
    names='species',
    title='Distribution of Iris Species (Count)',
    hole=0.3,
    color='species',
    color_discrete_map={
        'setosa': '#FF6B6B',
        'versicolor': '#4ECDC4',
        'virginica': "#9BD145"
    }
)

fig.update_traces(textinfo='label+percent+value')
fig.show()
```

```python
# Donut Chart of Average Sepal Width by Species

# First calculate mean sepal width per species
avg_data = df.groupby('species')['sepal width (cm)'].mean().reset_index()

fig = px.pie(
    avg_data,
    names='species',
    values='sepal width (cm)',
    title='Average Sepal Width by Species',
    hole=0.5,
    color='species',
    color_discrete_map={
        'setosa': '#FF6B6B',
        'versicolor': '#4ECDC4',
        'virginica': "#94D145"
    }
)

fig.update_traces(
    textinfo='label+percent+value',
    texttemplate='%{label}<br>%{value:.2f} cm'
)
fig.show()
```

```python
# Donut Chart of Total Petal Length by Species
fig = px.pie(
    df,
    names='species',
    values='petal length (cm)',
    title='Total Petal Length by Species',
    hole=0.2,
    color='species',
    color_discrete_map={
        'setosa': '#FF6B6B',
        'versicolor': '#4ECDC4',
        'virginica': "#9BD145"
    }
)

fig.update_traces(textinfo='label+percent+value')
fig.show()
```

### 7- Pyramid Chart :

```python
# Top 10 Countries Population in 2022 Pyramid 
df = pd.read_csv(url)
top10 = df.nlargest(10, '2022 Population')[['Country', '2022 Population']]

# For a population pyramid, we need age groups and male/female split
# Since this dataset doesn't have that, let's simulate or use another approach

#  Use actual country population as "Male" and a portion as "Female"

top10['Male'] = top10['2022 Population'] * 0.505  # ~50.5% male
top10['Female'] = top10['2022 Population'] * 0.495  # ~49.5% female

# Create the population pyramid
fig, ax = plt.subplots(figsize=(12, 8))

# Male bars (left side)
bars_male = ax.barh(
    top10['Country'],
    top10['Male'] / 1_000_000,  # Convert to millions
    color='skyblue',
    label='Male',
    edgecolor='black',
    height=0.6
)

# Female bars (right side)
bars_female = ax.barh(
    top10['Country'],
    -top10['Female'] / 1_000_000,  # Negative for left side
    color='pink',
    label='Female',
    edgecolor='black',
    height=0.6
)

# Add value labels
for bar in bars_male:
    width = bar.get_width()
    ax.text(width + 10, bar.get_y() + bar.get_height()/2,
            f'{width:.0f}M', ha='left', va='center', fontsize=9)

for bar in bars_female:
    width = abs(bar.get_width())
    ax.text(-width - 10, bar.get_y() + bar.get_height()/2,
            f'{width:.0f}M', ha='right', va='center', fontsize=9)

ax.set_xlabel('Population (Millions)', fontsize=12, fontweight='bold')
ax.set_title('Population Pyramid: Top 10 Countries (Male vs Female)\nSimulated Data for Demonstration',
             fontsize=14, fontweight='bold')
ax.legend(loc='lower right', fontsize=10)
ax.axvline(x=0, color='black', linewidth=1)
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

### 8-Funnel Chart :

Funnel charts are perfect for showing progressive stages where values decrease. 

```python
# Population by Continent

# Aggregate population by continent
continent_pop = df.groupby('Continent')['2022 Population'].sum().reset_index()
continent_pop = continent_pop.sort_values('2022 Population', ascending=False)

# Create funnel chart
fig = px.funnel(
    continent_pop,
    x='2022 Population',
    y='Continent',
    title='🌍 World Population Distribution by Continent (2022)',
    color='Continent',
    labels={'2022 Population': 'Total Population', 'Continent': 'Continent'}
)

fig.update_traces(
    textinfo='value+percent total',
    texttemplate='%{label}<br>%{value:,.0f} (%{percentTotal:.1%})',
    marker=dict(line=dict(width=1, color='white'))
)

fig.update_layout(width=800, height=600)
fig.show()
```

```python
# Top 10 Countries by Population in 2022

top10 = df.nlargest(10, '2022 Population')[['Country', '2022 Population']]
top10 = top10.sort_values('2022 Population', ascending=True)  # Ascending for funnel

fig = px.funnel(
    top10,
    x='2022 Population',
    y='Country',
    title='🔝 Top 10 Most Populous Countries (2022)',
    color='Country',
    labels={'2022 Population': 'Population', 'Country': 'Country'}
)

fig.update_traces(
    textinfo='value+percent total',
    texttemplate='%{label}<br>%{value:,.0f}',
    marker=dict(line=dict(width=1, color='white'))
)

fig.show()
```

```python
# Population Density Distribution

# Create density categories
def categorize_density(density):
    if density > 1000:
        return 'Very High (>1000)'
    elif density > 100:
        return 'High (100-1000)'
    elif density > 10:
        return 'Medium (10-100)'
    else:
        return 'Low (<10)'

df['Density Category'] = df['Density (per km²)'].apply(categorize_density)

# Count countries in each category
density_dist = df['Density Category'].value_counts().reset_index()
density_dist.columns = ['Density Category', 'Number of Countries']

# Define order
order = ['Very High (>1000)', 'High (100-1000)', 'Medium (10-100)', 'Low (<10)']
density_dist['Density Category'] = pd.Categorical(
    density_dist['Density Category'], 
    categories=order, 
    ordered=True
)
density_dist = density_dist.sort_values('Density Category')

fig = px.funnel(
    density_dist,
    x='Number of Countries',
    y='Density Category',
    title='🏙️ Distribution of Countries by Population Density',
    color='Density Category',
    labels={'Number of Countries': 'Count', 'Density Category': 'Category'}
)

fig.update_traces(
    textinfo='value+percent total',
    texttemplate='%{label}<br>%{value} countries (%{percentTotal:.1%})'
)

fig.show()
```

```python
# World Population Growth Funnel (2022 vs 2020 vs 2015)

# Show how population decreases as we go back in time
years = ['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population']
available_years = [y for y in years if y in df.columns]

if available_years:
    total_by_year = pd.DataFrame({
        'Year': [y.replace(' Population', '') for y in available_years],
        'Total Population': [df[y].sum() for y in available_years]
    })
    
    total_by_year = total_by_year.sort_values('Total Population', ascending=False)
    
    fig = px.funnel(
        total_by_year,
        x='Total Population',
        y='Year',
        title='📈 World Population Growth Over Time',
        labels={'Total Population': 'Total Population', 'Year': 'Year'}
    )
    
    fig.update_traces(
        textinfo='value+percent total',
        texttemplate='%{label}<br>%{value:,.0f}',
        marker=dict(
            color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'][:len(available_years)]
        )
    )
    
    fig.show()
```

```python
# Line Chart - Best for Population Growth ✅

import plotly.express as px
import pandas as pd


# Find all population year columns
year_cols = [col for col in df.columns if 'Population' in col and col != 'World Population Percentage']
year_cols = sorted(year_cols)

# Select top 5 countries
top5 = df.nlargest(5, '2022 Population')

# Create long format data for line chart
growth_data = []
for _, row in top5.iterrows():
    for year_col in year_cols:
        year = year_col.replace(' Population', '')
        growth_data.append({
            'Country': row['Country'],
            'Year': int(year),
            'Population': row[year_col]
        })

growth_df = pd.DataFrame(growth_data)

# Line chart
fig = px.line(
    growth_df,
    x='Year',
    y='Population',
    color='Country',
    title='📈 Population Growth: Top 5 Countries Over Time',
    markers=True,
    labels={'Population': 'Population', 'Year': 'Year'}
)

fig.update_layout(
    yaxis_tickformat=',',
    hovermode='x unified',
    height=600
)

fig.show()
```

### 9- Animated Bar Chart:

```python
# World Population Growth by Copuntry

year_cols = [col for col in df.columns if 'Population' in col and col != 'World Population Percentage']
year_cols = sorted(year_cols)

# Create long format
growth_data = []
for _, row in df.iterrows():
    for year_col in year_cols:
        year = year_col.replace(' Population', '')
        growth_data.append({
            'Country': row['Country'],
            'Continent': row['Continent'],
            'Year': int(year),
            'Population': row[year_col]
        })

growth_df = pd.DataFrame(growth_data)

# Animated bar chart
fig = px.bar(
    growth_df,
    x='Country',
    y='Population',
    color='Continent',
    animation_frame='Year',
    range_y=[0, growth_df['Population'].max() * 1.1],
    title='🌍 World Population Growth by Country (Animated)',
    labels={'Population': 'Population'}
)

fig.update_layout(
    xaxis={'categoryorder': 'total descending'},
    height=600
)

fig.show()
```

### 10- Area Chart:

```python
# Area Chart - Cumulative Growth

# Sum population by continent over years
continent_growth = []
for year_col in year_cols:
    year = year_col.replace(' Population', '')
    continent_totals = df.groupby('Continent')[year_col].sum()
    for continent, pop in continent_totals.items():
        continent_growth.append({
            'Continent': continent,
            'Year': int(year),
            'Population': pop
        })

continent_df = pd.DataFrame(continent_growth)

fig = px.area(
    continent_df,
    x='Year',
    y='Population',
    color='Continent',
    title='🏗️ Population Growth Stacked by Continent',
    labels={'Population': 'Total Population', 'Year': 'Year'},
    groupnorm=None
)

fig.update_layout(
    yaxis_tickformat=',',
    hovermode='x unified',
    height=600
)

fig.show()
```
